```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
train = pd.read_csv('train.csv')
movies = pd.read_csv('moviesData.csv')
```

```
values = movies.corr()
plt.figure(figsize=(10,10))
sns.heatmap(values,annot=True,vmin=-1,cmap='coolwarm')
plt.show()
```

```
# Creating a mask for removing duplicates 
mask = np.zeros_like(movies.corr(), dtype=np.bool) # create matrix of 0's of the given matrix site
mask[np.triu_indices_from(mask)] = True # changing the lower triangle matrix element to True (1)
plt.figure(figsize = (15,10))
sns.heatmap(values, vmin = -1, cmap="coolwarm", annot=True, mask=mask)
plt.show()
```

Density Plot

```
from pylab import * 
df = sns.load_dataset('iris')
subplot(2,2,1)
# density plot with shade
sns.kdeplot(df['sepal_width'], shade=True, color='r')

subplot(2,2,2)

sns.kdeplot(df['sepal_length'], shade=True, color='b')

subplot(2,2,3)

sns.kdeplot(df['petal_width'], shade=True, color='g')

subplot(2,2,4)

sns.kdeplot(df['petal_length'], shade=True, color='y')
```

```
from pylab import * 

subplot(2,2,1)
# density plot with shade
sns.kdeplot(df['sepal_width'], shade=True, color='r',cumulative=True)

subplot(2,2,2)

sns.kdeplot(df['sepal_length'], shade=True, color='b',cumulative=True)

subplot(2,2,3)

sns.kdeplot(df['petal_width'], shade=True, color='g',cumulative=True)

subplot(2,2,4)

sns.kdeplot(df['petal_length'], shade=True, color='y',cumulative=True)
```
